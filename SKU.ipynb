```python
df = pd.DataFrame(df.T)
df.head()
```

```python
df.columns = df.iloc[0]
df.head()
```

```python
df = df.iloc[1:]
df.head()
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df = df.reset_index()
df.head()
```

```python
df.rename(columns = {'index' : 'Date'}, inplace=True)
```

```python
df.dtypes
```

```python
df['Month'] = df.Date.dt.month
df.head()
```

```python
df.groupby('Month').sum()
```

```python
df['Quarter'] = df.Date.dt.quarter
df.head()
```

```python
df1 = df.copy()
df1.drop('Month', axis=1, inplace=True)
```

```python
df1.groupby('Quarter').sum()
```

```python
df2 = df.copy()
df2.drop('Quarter', axis=1, inplace=True)
```

```python
df2 = df2.groupby('Month').sum()
df2.head()
```

```python
#Top three months:-
df2.sum(axis=1).sort_values(ascending=False)[:3]
```

```python
df2.describe()
```

```python
Q1 = df2.quantile(0.25)
Q3 = df2.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
df_outliers = pd.DataFrame((df2 < (Q1 - 1.5 * IQR)) | (df2 > (Q3 + 1.5 * IQR)))
df_outliers.sum()
```

```python
df3 = pd.DataFrame(df_outliers.sum())
df3.rename(columns={0:'outliers'}, inplace=True)
df3
```

```python
df3.loc[df3.outliers > 0]
```
